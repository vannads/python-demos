```python
from gensim.summarization import summarize
import sys
from os import path

def myfunction(fname):
    fname = sys.argv[1]
    myStr=""
    with open(fname, 'r') as f:
        content = f.read()
        summary = summarize(content, split=True)
        for i, sentence in enumerate(summary):
            #print("%d) %s" % (i+1, sentence))
            myStr += sentence
    return myStr

```
