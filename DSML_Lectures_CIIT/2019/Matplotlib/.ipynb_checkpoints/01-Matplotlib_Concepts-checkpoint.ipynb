```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Example: 1

plt.plot(x, y, 'b')
plt.xlabel('X-axis')
plt.ylabel('Y-Axis')
plt.title('Graph-1')
```

```python
# Example: 2

plt.plot(x, y, 'r')
plt.xlabel('X-axis')
plt.ylabel('Y-Axis')
plt.title('Graph-1')
plt.show()
```

- Subplots

```python
# Example: 3

plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g.-')
plt.show()
```

- OO Method

```python
# Example: 4

# Create a figure, or an empty canvas
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'b')
axes.set_xlabel("X-axis")
axes.set_ylabel("Y-axis")
axes.set_title("Graph-3")

# axes.show() -- Avoid this
```

```python
# Example: 5

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.6])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Fig1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('Fig1')

# Fig2

axes2.plot(x, y, 'g')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
axes2.set_title('Fig2')
```

```python
# Example: 5

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.6])
axes2 = fig.add_axes([0.2, 0.4, 0.4, 0.2])

# Fig1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('Fig1')

# Fig2

axes2.plot(x, y, 'g')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
axes2.set_title('Fig2')
```

```python
# Example: 6

fig = plt.figure()

axes1 = fig.add_axes([0, 0, 0.8, 0.8])
axes2 = fig.add_axes([0.1, 0.4, 0.4, 0.3])

# Fig1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('Fig1')

# Fig2

axes2.plot(x, y, 'g')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
axes2.set_title('Fig2')
```

```python
# Example: 7

# fig = plt.figure()
fig, ax = plt.subplots()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2,'b.-', label='x**2')
ax.plot(x, x**3,'g--', label='x**3')

ax.legend(loc=0)
```
